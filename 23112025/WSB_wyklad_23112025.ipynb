```python
temperaturaFahr = 77

temperaturaCelc = (temperaturaFahr - 32) * 5 / 9

print(temperaturaCelc)
```

```python
temperaturaFahr2 = 111

temperaturaCelc2 = (temperaturaFahr2 - 32) * 5 / 9

print(temperaturaCelc2)
```

```python
temperaturaFahr3 = 90

temperaturaCelc3 = (temperaturaFahr3 - 32) * 5 / 9

print(temperaturaCelc3)
```

```python
def fahr_to_celc(fahr):
  celc = (fahr - 32) * 5 / 9
  return celc
```

```python
temperaturaFahr4 = 81

temperaturaCelc4 = fahr_to_celc(temperaturaFahr4)

print(temperaturaCelc4)
```
