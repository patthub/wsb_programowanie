```python
#3. Napisz program, który przyjmie listę [1, 2, 3, 4, 5]
#i w wyniku zwróci wartości pomnożone przez liczbę podaną przez użytkownika (w inpucie)


lista = [1, 2, 3, 4, 5]

mnoznik = input("Podaj liczbe: ")

if mnoznik.isdigit() == True:
  mnoznik_int = int(mnoznik)

for liczba in lista:
  print(liczba * mnoznik_int)
```

```python
#5. Napisz program, który poda sumę wszystkich liczb w liście [1, 2, 3, 4, 5]
#UZYWAJAC PETLI FOR
lista = [1, 2, 3, 4, 5]

licznik = 0

for elem in lista:
  licznik = licznik + elem #inkrementacja
```
