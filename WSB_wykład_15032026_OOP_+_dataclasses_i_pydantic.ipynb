```python
#NADPISYWANIE METOD

class Animal:
    def speak(self):
        return '...'


class Dog(Animal):
    def speak(self):
        return 'Hau!'


class Cat(Animal):
    def speak(self):
        return 'Miau!'


class Fish(Animal):
    pass


# polimorfizm - ta sama pętla, różne zachowanie
animals = [Dog(), Cat(), Fish()]
for animal in animals:
    print(f'{animal.__class__.__name__}: {animal.speak()}')
```

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        return '...'

    def describe(self):
        return f'Jestem {self.name}, mam {self.age} lat. Mówię: {self.speak()}'


class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.breed = breed

    def speak(self):
        return 'Hau!'


class Cat(Animal):
    def __init__(self, name, age, indoor=True):
        super().__init__(name, age)
        self.indoor = indoor

    def speak(self):
        return 'Miau!'



rex = Dog('Rex', 3, 'Owczarek')


whiskers = Cat('Wąsik', 5)

print(rex.describe())
print(whiskers.describe())

print(rex.breed)
print(whiskers.indoor)

```

```python

class Vehicle:
    def describe(self):
        return 'Jestem pojazdem'


class Car(Vehicle):
    def describe(self):
        base = super().describe()
        return base + ', konkretnie samochodem'


class ElectricCar(Car):
    def describe(self):
        base = super().describe()
        return base + ', elektrycznym'

print(Vehicle().describe())
print(Car().describe())
print(ElectricCar().describe())
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        print(f'  Person: {name}, {age}')


class Student(Person):
    def __init__(self, name, age, major):
        super().__init__(name, age)
        self.major = major
        print(f'  Student: kierunek {major}')


class PhDStudent(Student):
    def __init__(self, name, age, major, thesis_topic):
        super().__init__(name, age, major)
        self.thesis_topic = thesis_topic
        print(f'  PhDStudent: temat {thesis_topic}')


print('Tworzenie PhDStudent:')
phd = PhDStudent('Anna', 28, 'Informatyka', 'Sieci neuronowe')
print(phd.name)
print(phd.major)
print(phd.thesis_topic)
```

```python
#METODY DUNDER - double under

class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f'{self.name} wydaje dźwięk'

    def __str__(self):
        return f'{self.name}'

    def __repr__(self):
        return f"Animal(name={self.name!r})"

rex = Animal('Rex')
print(rex)
```

```python
# dunder methods są wszędzie — używamy ich od początku
print(1 + 2)         # int.__add__
print(len([1,2,3]))  # list.__len__
print('ha' * 3)      # str.__mul__

         # Rex              — __str__
print(repr(rex))     # Animal(name='Rex') — __repr__

animals = [Animal('Rex'), Animal('Burek')]
print(animals)       # lista używa __repr__ każdego elementu
```

```python
#PRZED DATACLASS

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f'Point(x={self.x}, y={self.y})'

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y


p1 = Point(1, 2)
p2 = Point(1, 2)
print(p1)
print(p1 == p2)
```

```python
#PO DATACLASS

from dataclasses import dataclass, field
from abc import ABC, abstractmethod
from typing import Optional



class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

@dataclass #dekorator
class Point:
    x: float
    y: float


p1 = Point(1, 2)
p2 = Point(1, 2)
print(p1)           # Point(x=1, y=2)  — __repr__ gratis
print(p1 == p2)     # True             — __eq__ gratis


@dataclass
class Product:
    name: str
    price: float
    in_stock: bool = True


p1 = Product('Laptop', 3999.99)
p2 = Product('Mysz', 49.99, in_stock=False)
print(p1)
print(p2)
```

```python
# ---- 3. field() — mutable defaults ----

@dataclass
class Cart:
    owner: str
    items: list = field(default_factory=list)   # każda instancja dostaje nową listę


cart1 = Cart('Anna')
cart2 = Cart('Jan')

cart1.items.append('Laptop')

print(cart1.items)   # ['Laptop']
print(cart2.items)   # []  — osobna lista!


@dataclass
class Employee:
    name: str
    salary: float
    bonus: float = field(default=0.0)
    id_number: int = field(init=False, default=0)   # nie przekazujemy w konstruktorze


e = Employee('Anna', 5000)
print(e)
```

```python
# ---- 4. __post_init__ ----

@dataclass
class Rectangle:
    width: float
    height: float
    area: float = field(init=False)

    def __post_init__(self):
        if self.width <= 0 or self.height <= 0:
            raise ValueError('Wymiary muszą być dodatnie')
        self.area = self.width * self.height


r = Rectangle(4, 5)
print(r)
print(r.area)

try:
    bad = Rectangle(-1, 5)
except ValueError as e:
    print(f'Błąd: {e}')
```

```python
# ---- 5. frozen=True ----

@dataclass(frozen=True)
class Config:
    host: str
    port: int = 8080
    debug: bool = False


cfg = Config('localhost', port=5000)
print(cfg)

try:
    cfg.host = 'inny-serwer'
except Exception as e:
    print(f'Błąd: {e}')

cache = {cfg: 'wynik zapytania'}
print(cache[cfg])
```

```python
# ---- 6. ABC ----

class Shape(ABC):

    @abstractmethod
    def area(self) -> float:
        pass

    @abstractmethod
    def perimeter(self) -> float:
        pass

    def describe(self) -> str:          # wspólna metoda — nie trzeba nadpisywać
        return f'Pole: {self.area():.2f}, obwód: {self.perimeter():.2f}'


try:
    s = Shape()
except TypeError as e:
    print(f'Błąd: {e}')
```

```python
# ---- 7. IMPLEMENTACJE ABC ----

#klasy konkretne -- takie klasy które mogą mieć swoje obiekty i są domyślnymi typami class w Pythonie
#klasy abstrakcyjne -- klasy które nie mają swoich obiektów i służą do dziedziczenia

class Circle(Shape):
    def __init__(self, radius: float):
        self.radius = radius

    def area(self) -> float:
        import math
        return math.pi * self.radius ** 2

    def perimeter(self) -> float:
        import math
        return 2 * math.pi * self.radius


class Rectangle(Shape):
    def __init__(self, width: float, height: float):
        self.width = width
        self.height = height

    def area(self) -> float:
        return self.width * self.height

    def perimeter(self) -> float:
        return 2 * (self.width + self.height)


shapes = [Circle(5), Rectangle(4, 6)]
for shape in shapes:
    print(f'{shape.__class__.__name__}: {shape.describe()}')


# brakująca metoda — błąd przy tworzeniu obiektu
class IncompleteShape(Shape):
    def area(self):
        return 0

try:
    s = IncompleteShape()
except TypeError as e:
    print(f'Błąd: {e}')
```

```python
# ---- 8. DATACLASS + ABC RAZEM ----

@dataclass
class Animal(ABC):
    name: str
    age: int

    @abstractmethod
    def speak(self) -> str:
        pass

    def introduce(self) -> str:
        return f'Jestem {self.name}, mam {self.age} lat. Mówię: {self.speak()}'


@dataclass
class Dog(Animal):
    breed: str = 'nieznana'

    def speak(self) -> str:
        return 'Hau!'


@dataclass
class Cat(Animal):
    indoor: bool = True

    def speak(self) -> str:
        return 'Miau!'


rex = Dog('Rex', 3, breed='Owczarek')
whiskers = Cat('Wąsik', 5, indoor=True)

print(rex)
print(whiskers)
print(rex.introduce())
print(whiskers.introduce())
```
