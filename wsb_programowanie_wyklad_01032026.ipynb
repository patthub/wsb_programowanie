```python
class Student:
  #konstruktor (klasy)
  def __init__(self, imie, nazwisko, ocena):
    self.imie = imie
    self.nazwisko = nazwisko
    self.ocena = ocena

  def przedstaw_sie(self):
    wynik = f"Nazywam się {self.imie} {self.nazwisko} i mam ocenę {self.ocena}"
    return wynik


```

```python
class Osoba:
  #konstruktor
  def __init__(self, imie, nazwisko):
    self.imie = imie
    self.nazwisko = nazwisko


class Student(Osoba):
  #konstruktor (klasy)
  def __init__(self, imie, nazwisko, ocena):
    super().__init__(imie, nazwisko)
    self.ocena = ocena

  def przedstaw_sie(self):
    wynik = f"Nazywam się {self.imie} {self.nazwisko} i mam ocenę {self.ocena}"
    return wynik


class Wykladowca(Osoba):
  #konstruktor
  def __init__(self, imie, nazwisko, wyplata):
    super().__init__(imie, nazwisko)
    self.wyplata = wyplata

  def ocen(self):
    pass
```
